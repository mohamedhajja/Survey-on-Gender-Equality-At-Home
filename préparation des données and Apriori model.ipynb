<h2>Apriori Algoritm</h2>

<h3>Importing the libraries</h3>

```python
import pandas as pd 
import numpy as np
from mlxtend.preprocessing import TransactionEncoder
import statistics as st
from mlxtend.frequent_patterns import apriori,association_rules
```

<h3>Data Preprocessing</h3>

```python
data=pd.read_excel("sog_agg_country.xlsx")
```

```python
data.head()
```

```python
data.columns
```

```python
list=[]
for i in range(87):
    for j in data.index:
        if( data.iloc[j,i]>=data[data.columns[i]].mean()):
                list.append(data.columns[i])
        else:
                list.append(np.nan)
```

```python
dim_list=[]
list_Fin=[]
for i in range(87):
    list_Fin.append(dim_list)
```

```python
len(list)
```

```python
k=0
j=0
for i in range(0,21402,246):
    j=j+246
    list_Fin[k]=list[i:j]
    k=k+1
```

```python
len(list_Fin[1])==len(data[data.columns[1]])
```

```python
df=pd.DataFrame(list_Fin)
```

```python
df.head()
```

```python
df=df.fillna(value='HAJJA')
```

```python
df.head()
```

```python
lalist=[]
for i in range(246 ):
    lalist.append(df[i].tolist())
```

```python
te=TransactionEncoder()
tr_wary=te.fit(lalist).transform(lalist)
```

```python
moha=pd.DataFrame(tr_wary ,columns=te.columns_)
```

```python
moha.head()
```

```python
moha.drop('HAJJA',axis=1,inplace=True)
```

<h3>Trainng the Apriori model on the dataset</h3>

```python
freq=apriori(moha,min_support=0.4,use_colnames=True)
rules=association_rules(freq,metric='lift',min_threshold=1)
rules=rules.sort_values(['confidence','lift'],ascending=[False,False])
```

<h3>results</h3>

```python
print(rules)
```
